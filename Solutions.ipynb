Find the sum of all elements in a list.

```python
numbers=[34,56,78,90,99,78,34]
total=0
for i in numbers:
    total+= i
print(total)
```

Find maximum and minimum number in a list

```python
a=[23,45,6,78,9,34,56,67,98]
small=a[0]
max=a[0]
for n in a:
    if small<n:
        small=n
    if max>n:
        max=n
print(small,max)
    
```

Count occurrences of a specific number

```python
a=[23,4,5,6,4,5,4,6,3,5,4]
num=4
count=0
for n in a :
    if n==num:
        count+=1
print(count)
```

Remove all occurrences of an element

```python
a=[5,6,7,5,5,64,5,3,5,6,3,2]
num=5
result=[]
for i in a:
    if i !=num:
        result.append(i)
print(result)
```

Reverse a list without using reverse()

```python
a=[90,56,98,45,23,12,35,67,90]
rev=a[::-1]  #slicing 
print(rev)
```

```python
#using loop 
a=[90,56,98,45,23,12,35,67,90]
rev=[]
for i in a:
    rev.insert(0,i)
print(rev)
```

Check if list is sorted

```python
nums=[89,34,90,23,456,87,93,125,10]
sorted_flag=True
for i in range(len(nums)-1):
    if nums[i]>nums[i+1]:
        sorted_flag=False
        break
print(sorted_flag)
```
